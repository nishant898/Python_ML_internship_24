```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns 
import os
%matplotlib inline
```

```python
auto =pd.read_csv("titanic.csv")
auto.head()
```

```python
sexVSalive=auto.groupby(["sex","alive"]).size().unstack(fill_value=0)
sexVSalive
```

```python
sexVSalive.plot(kind='bar',stacked=True)
plt.show()
```

```python
plt.figure(figsize=(20,8))

sns.lineplot(data=auto["alive"],linewidth=2,label="alive")
sns.lineplot(data=auto["adult_male"],linewidth=2,label="adult_male")
plt.title("alive VS adult_male")
plt.xlabel("alive")
plt.ylabel("adult_male")

plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.set(rc={"axes.facecolor":"#FFFFFF","axes.grid":False ,"xtick.labelsize":10, "ytick.labelsize":10})
plt.title("class VS who")
plt.xlabel("class")
plt.ylabel("who")
plt.xticks(rotation=45)

sns.lineplot(x=auto["age"],y=auto["who"] . index.values,color='r')
plt.show()
```
