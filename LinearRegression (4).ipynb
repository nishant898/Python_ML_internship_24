```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression as LR
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
#linear regression 
#single class
#engine_size vs prize
#plot the graph
#error find 
```

```python
np.random.seed(0)
E_S=np.random.normal(2.0,0.5,100)
price= 1000+E_S*1000 +np.random.normal(0,2000,100)
```

```python
E_S
```

```python
price
```

```python
df = pd.DataFrame({"Engine Size":E_S,"Price":price})
df
```

```python
X=df[["Engine Size"]]
X
```

```python
Y=df["Price"]
Y
```

```python
model = LR()
model.fit(X,Y)
```

```python
y=model.predict(X)
y
```

```python
MSE=mean_squared_error(Y,y)
MAE=mean_absolute_error(Y,y)
R2=r2_score(Y,y)
```

```python
R2
```

```python
print(f"Mean Squared Error = {MSE}\nMean absolute error ={MAE}\nR2=score={R2}")
```

```python
plt.figure(figure=(10,5))
plt.scatter(X,Y,color='blue',label="Actual Data")
plt.plot(X,y,color='red',label="Predicted Data")
plt.xlabel("Engine size")
plt.ylabel("Price")
plt.title("Linear Regression")
plt.legend()
plt.show()
```

```python
np.random.seed(0)
E_S=np.random.normal(3.0,0.5,100)
W_b=np.random.normal(2.0,0.5,100)
amt= 1000+E_S*1000 +np.random.normal(0,2000,100)
```

```python
print(E_S)
print(W_b)
print(amt)
```

```python
df2 = pd.DataFrame({"Engine Size":E_S,"Wheel Base":W_b,"Amount":amt})
df2
```

```python
X1 = df2[["Engine Size","Wheel Base"]]
Y1 = df2["Amount"]
```

```python
MLRModel= LR()
MLRModel.fit(X1,Y1)
```

```python
y1=MLRModel.predict(X1)
```

```python
MSE1=mean_squared_error(Y1,y1)
MAE1=mean_absolute_error(Y1,y1)
R21=r2_score(Y1,y1)
print(f"Mean squared Error = {MSE1}\n Mean absolute Error = {MAE1} \n R2 score = {R21}")
```

```python
x_surf,y_surf = np.meshgrid(
    np.linspace(
        df2['Engine Size'].min(),
        df2['Engine Size'].max(),
        100
    ),
    np.linspace(
        df2['Wheel Base'].min(),
        df2['Wheel Base'].max(),100))
```

```python
x_surf
```

```python
y_surf
```

```python
xy_surf=np.c_[x_surf.ravel(),y_surf.ravel()]
xy_surf
```

```python
z_surf=MLRModel.predict(xy_surf).reshape(x_surf.shape)
```

```python
z_surf
```

```python
fig = plt.figure()
ax= fig.add_subplot(111,projection ='3d')
ax.scatter(df["Engine Size"],df2["Wheel Base"],df2["Amount"],color='blue',label="Actual data")
ax.plot_surface(,y_surf,z_surf,color="red",label= "Predicted Value",alpha=0.5)
ax.set_xlabel("Engine Size")
ax.set_ylabel("Wheel Base")
ax.set_zlabel("Price")
ax.set_title("Engine Size and Wheel Base VS price")
plt.legend()
plt.show()
```
